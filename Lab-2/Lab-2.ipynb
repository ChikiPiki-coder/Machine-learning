```
#max=data.groupby("weather")["temp_max"].mean()
#min=data.groupby("weather")["temp_min"].mean()
#max.plot(kind='bar', figsize=(10,6), color='red')
#min.plot(kind='bar', figsize=(10,6))

result = data.groupby(["owner"])['year'].mean()
result.plot(kind='bar', figsize=(10,6), color='blue') #formerly: sns.barplot(x='Id', y="Speed", data=df, palette=colors, order=result['Id'])
#plt.show()
```

```
fig, (axis1) = plt.subplots(1,figsize=(15,5))
sns.countplot(x='owner',data=data,palette="Blues_d") # график количества владельцев машин
```

```
plt.figure(figsize=(15,10)) #зависимость стоимости машины от продавца
sns.boxplot(x='selling_price', y='seller_type',data=data,palette="magma")
```

```
type_counts = data['transmission'].value_counts() #диаграмма о коробке передач 
df2 = pd.DataFrame({'type': type_counts}, index = ['Manual', 'Automatic'])
mycolors = ["CadetBlue", "Cornsilk"]
df2.plot.pie(y='type',shadow = True, figsize=(10,10), autopct='%1.1f%%', colors = mycolors)
```

```
#sns.set(style="darkgrid") #график о зависимости количества владельцев от пробега
sns.catplot(x='owner', #зависимость типа карты от возраста
            y='km_driven',
            data=data,  height=8.27, aspect=12.7/8.27)
```

```
fig,axs=plt.subplots(2,1,figsize=(15,10)) #гистограммы о стоимости и пробеге машины
sns.histplot(data=data,x="selling_price",kde=True,ax=axs[0],color='green')
sns.histplot(data=data,x="km_driven",kde=True,ax=axs[1],color='red')
```

```
colors_list = ['DarkGreen', 'DarkOliveGreen',  'DarkSeaGreen',  'LightGreen',  'LimeGreen']

plt.figure(figsize=(15,10))
sns.violinplot(x='fuel', y='km_driven',palette=colors_list, data=data)
#sns.swarmplot(x='fuel', y='km_driven', color="k", alpha=0.8, data=data)
plt.title("Fuel")
plt.ylabel("Mieage")
```
